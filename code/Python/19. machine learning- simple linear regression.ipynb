# Regression models

### simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn

from pylab import rcParams
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize']=10,8
```

```python
rooms = 2*np.random.rand(100,1)+3
rooms[1:10]
```

```python
price = 265+6*rooms+abs(np.random.rand(100,1))
price[1:10]
```

```python
plt.plot(rooms,price,'r^')
plt.xlabel('# of Rooms,2019 Average')
plt.ylabel('2019 Average Home,1000s USD')
plt.show()
```

```python
plt.plot(rooms,price)
plt.show()
```

```python
x = rooms
y = price

LinReg = LinearRegression()
LinReg.fit(x,y)
print(LinReg.intercept_, LinReg.coef_)
```

Simple Algebra
- y = mx + b
- b = intercept = 265.65
Estimated Coefficients
- LinReg.coef_ = [5.96] Estimated coefficients for the terms in the linear regression problem.

```python
print(LinReg.score(x,y)) # return coefficient of determination which is the r squared of the prediction
```
