```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize']=8,4
```

```python
x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]
plt.grid()
plt.bar(x,y)

plt.xlabel(' your x axis label')
plt.ylabel('your y axies label')
```

```python
# labeling plot features
plt.xlabel(' your x axis label')
plt.ylabel('your y axies label')
```

```python
z = [1,2,3,4,.5]
veh_type = ['bicycle','motorbike','car','van','stroller']

plt.pie(z,labels=veh_type)
plt.show()
```

```python
# The object-oriented method
cars = pd.read_csv('mtcars.csv')
cars.columns = ['car_name','mpg','cyl','disp','hp','drap','wt','qsec','vs','am','gear','carb']
mpg = cars.mpg # select the mpg values

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_name,rotation=60,fontsize='medium')# no xticks,rotation label
ax.set_title('Miles per gallon of cars in mtcars dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')
```

```python
# adding a legend to your plot

plt.pie(z)
plt.legend(veh_type,loc='best')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1]) # add an axes to the figure.

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_name,rotation=60, fontsize='medium')
ax.set_title('miles per gallon of cars in mtcars dataset')
ax.set_xlabel('car name')
ax.set_ylabel('miles/gal')
ax.legend(loc='best')
ax.grid()
```
