```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
#Creating a line chart from a list object

#plot a line chart in matplotlib
x = range(1,10)
y = [1,2,3,4,0,5,3,2,1]

plt.plot(x,y)
```

```python
# plot a line chart from Pandas object
```

```python
cars = pd.read_csv('mtcars.csv')
cars.columns = ['car_name','mpg','cyl','disp','hp','drap','wt','qsec','vs','am','gear','carb']
mpg = cars['mpg']
mpg.plot()
```

```python
df = cars[['mpg','wt','hp']]
df.plot()
```

```python
#creating bar charts

#creating a bar chart from a list
plt.bar(x,y)
```

```python
# creating a bar chart from Pandas objects
mpg.plot(kind="bar")
```

```python
mpg.plot(kind='barh')
```
