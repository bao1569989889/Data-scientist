```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series,DataFrame

from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

```python
import seaborn as sb
sb.set_style('whitegrid')
```

```python
cars = pd.read_csv('/Users/tank/Desktop/python/linkedin_python_Exercise Files/Data/mtcars.csv')
cars.columns = ['car_name','mpg','cyl','disp','hp','drap','wt','qsec','vs','am','gear','carb']
cars.index = cars.car_name
mpg=cars['mpg']
mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sb.distplot(mpg)
```

```python
cars.plot(kind='scatter',x='hp',y='mpg',c=['green'],s=150)# c: color; s:size
```

```python
sb.regplot(x='hp',y='mpg',data=cars,scatter=True)
```

```python
sb.regplot(x='hp',y='mpg',data=cars,scatter=False)
```

```python
# generating a scatter plot matrix
sb.pairplot(cars)
```

```python
cars_subset =cars[['mpg','disp','hp','wt']]
sb.pairplot(cars_subset)
plt.show()
```

```python
cars_subset =cars[['mpg','disp']]
sb.pairplot(cars_subset)
plt.show()
```

```python
# Boxplots

cars.boxplot(column='mpg',by='am')
cars.boxplot(column='wt',by='am')
```

```python
sb.boxplot(x='am',y='mpg',data=cars,palette='hls')
```
